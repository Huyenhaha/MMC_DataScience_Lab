```python
def analyzeClass():
    while True:
        filename = input()
        FILE_PATH = './Data_Files/' + filename + '.txt'
        try:
            f = open(FILE_PATH,'r+')
        except Exception as e:
            print(e)
        else:
            print('\n SUCCESSFULLY OPENED\n', filename)
          #  f.close()
          #  break
    
        print("**** ANALYZING ****")
    
        k = 0 # number of lines
        error = 0 # number of lines error
        
        for line in f.readlines():
            k += 1
            listValue = line.rstrip('\n').split(',')
            if(len(listValue) != 26):
                print('invalid line of data \n')
                print(line,'\n')
                error += 1
            elif not listValue[0][1:].isnumeric() or len(listValue[0]) != 9:
                print('Invalid line of data : N# is invalid \n')
                print(line, '\n')
                error += 1
        if error == 0:
            print('no error found!')
        
        print('\n**** Report ****\n')
        print('số dòng hợp lệ : ', k - error, '\n')
        print('số dòng bị lỗi:' , error)
        
        choice = input('bạn có muốn nhập thêm không? (0: không, 1: có)')
        if int(choice) == 0:
            break
```

```python
while True:
    filename = input()
    FILE_PATH = './Data_Files/' + filename + '.txt'
    try:
        f = open(FILE_PATH,'r+')
    except Exception as e:
        print(e)
    else:
        print('\n SUCCESSFULLY OPENED\n', filename)
      #  f.close()
      #  break

    print("**** ANALYZING ****")

    k = 0 # number of lines
    error = 0 # number of lines error
    grade = [] # list grade
    valid_line = []
    answer_key = 'B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D'.split(',')
    for line in f.readlines():
        k += 1
        listValue = line.rstrip('\n').split(',')
        if(len(listValue) != 26):
            print('invalid line of data \n')
            print(line,'\n')
            error += 1
        elif not listValue[0][1:].isnumeric() or len(listValue[0]) != 9:
            print('Invalid line of data : N# is invalid \n')
            print(line, '\n')
            error += 1
        else:
            valid_line.append(listValue)
    print(valid_line[1])
    for line in valid_line:
        answer = line[1:]
        score = 0
        for j in range(25):
            if answer[j] == answer_key[j] :
                score += 4
            elif answer[j] =='':
                score += 0
            else :
                score -= 1
        grade.append(score)

    if error == 0:
        print('no error found!')
        
    print('\n**** Report ****\n')
    print('số dòng hợp lệ : ', k - error, '\n')
    print('số dòng bị lỗi:' , error)
    print('Mean (average) score : ', sum(grade)/len(grade),'\n')
    print('Highest score : ', max(grade), '\n')
    print('Lowest score: ', min(grade), '\n')
    print('Range of scores: ', max(grade) - min(grade), '\n')
    s = sorted(grade)
    n = len(grade)
    if n % 2 == 0:
        median = (s[n//2 - 1] + s[n//2]) /2.0
    else:
        median = s[n//2]
    print('Median score: ', median, '\n')
    
    choice = input('bạn có muốn nhập thêm không? (0: không, 1: có)')
    if int(choice) == 0:
        break

    
```

# Task 4

```python
while True:
    filename = input()
    FILE_PATH = './Data_Files/' + filename + '.txt'
    try:
        f = open(FILE_PATH,'r+')
    except Exception as e:
        print(e)
    else:
        print('\n SUCCESSFULLY OPENED\n', filename)
      #  f.close()
      #  break

    print("**** ANALYZING ****")

    student_number = []
    grade = [] # list grade
    valid_line = []
    answer_key = 'B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D'.split(',')
    for line in f.readlines():
        listValue = line.rstrip('\n').split(',')
        if len(listValue) == 26 and listValue[0][1:].isnumeric() and len(listValue[0]) == 9:
            valid_line.append(listValue)
            
    for line in valid_line:
        student_number.append(line[0])
        answer = line[1:]
        score = 0
        for j in range(25):
            if answer[j] == answer_key[j] :
                score += 4
            elif answer[j] =='':
                score += 0
            else :
                score -= 1
        grade.append(score)
    FILE_PATH_GRADE = './Data_Files/' + filename + '_grade.txt'
    with open(FILE_PATH_GRADE, 'w') as fw:
         for sn,gr in zip(student_number,grade):
             fw.write(sn + ',' + str(gr) + '\n')
         print('successfully write file ' + filename + "_grade.txt \n")
    f.close()
    
    choice = input('bạn có muốn nhập thêm không? (0: không, 1: có)')
    if int(choice) == 0:
        break

    
```
